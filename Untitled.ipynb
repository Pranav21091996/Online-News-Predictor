```python
import pandas as pd
import numpy as np
from utils import *
data = pd.read_csv('OnlineNewsPopularity.csv')  #importing xlsx file
```

```python
data.head(5)
```

```python
variables = list(data)
variables = variables[2:]
print(variables)            #removing serial number column and getting list of dependent and independent variables
```

```python
dataset = {}
for parameters in variables:
    dataset[parameters]=data[parameters].values.tolist() #dictionary of the list of values of all the parameters
```

```python
for parameters in variables:
    dataset[parameters]=(dataset[parameters]-np.min(dataset[parameters]))/(np.max(dataset[parameters])-np.min(dataset[parameters]))
```

```python
pd.DataFrame.from_dict(dataset)
```

```python
for i in range(len(dataset[' shares'])):
    if (dataset[' shares'][i]>1400):
        dataset[' shares'][i]=1
    else:
        dataset[' shares'][i]=0
```

```python
# creating train and test data for the neural network

input_data = []
input_label = []

for i in range(len(data)):
    list = []
    for parameters in variables:
        if(parameters!=' shares'):
            list.append(dataset[parameters][i])
    input_data.append(list)    
    input_label.append(dataset[' shares'][i])
```

```python
encoder = LabelEncoder()
encoder.fit(input_label)
encoded_Y = encoder.transform(input_label)
```

```python
np.shape(input_data)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(np.array(input_data), encoded_Y, test_size=0.2)

X_train, X_train_lr, y_train, y_train_lr = train_test_split(X_train,
                                                            y_train,
                                                            test_size=0.2)

from keras.wrappers.scikit_learn import KerasClassifier
keras_model = create_baseline()
keras_model.fit(X_train, y_train, epochs=200, batch_size=64, verbose=0)

from sklearn.metrics import roc_curve
y_pred_keras = keras_model.predict(X_test).ravel()
fpr_keras, tpr_keras, thresholds_keras = roc_curve(y_test, y_pred_keras)
```

```python
from sklearn.metrics import auc
auc_keras = auc(fpr_keras, tpr_keras)
```

```python
from sklearn.ensemble import RandomForestClassifier
# Supervised transformation based on random forests
rf = RandomForestClassifier(max_depth=3, n_estimators=10)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict_proba(X_test)[:, 1]
fpr_rf, tpr_rf, thresholds_rf = roc_curve(y_test, y_pred_rf)
auc_rf = auc(fpr_rf, tpr_rf)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(1)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr_keras, tpr_keras, label='NN (area = {:.3f})'.format(auc_keras))
plt.plot(fpr_rf, tpr_rf, label='RF (area = {:.3f})'.format(auc_rf))
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve')
plt.legend(loc='best')
plt.show()
```
